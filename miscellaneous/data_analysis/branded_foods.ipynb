## Ładownie danych do Data Frame'a

### 1 - wybór produktów *branded* i dołączenie podstawowych pól

- Kolumna `data_type` w `food.csv` jest normalizowana (zamiana na małe litery, usunięcie zbędnych spacji), a następnie wybierane są wyłącznie rekordy typu `branded_food` lub `branded`.
- Z przefiltrowanej tabeli pozostawiane są kolumny `fdc_id`, `description` oraz `publication_date`.
- Do tych rekordów dołączane są dane z `branded_food.csv` (po `fdc_id`), w szczególności:
  - `brand_owner`, `brand_name`
  - `serving_size`, `serving_size_unit`
  - `household_serving_fulltext`
  - `gtin_upc`
- Wynik jest zapisywany w zmiennej `food_b` jako bazowa tabela produktów markowych z identyfikatorem USDA, nazwą, datą publikacji oraz informacjami o producencie, porcji i kodzie kreskowym.


````python
import pandas as pd
import re

# wybieramy tylko branded_food z food.csv
food_fix = food.copy()
food_fix["data_type_l"] = food_fix["data_type"].str.lower().str.strip()

food_b = (
    food_fix.loc[
        food_fix["data_type_l"].isin(["branded_food", "branded"]),
        ["fdc_id", "description", "publication_date"]
    ]
    .copy()
)

print("Branded-only (po filtrze):", food_b.shape)

# dokładamy pola z branded_food.csv, w tym brand_owner / brand_name / gtin_upc
food_b = food_b.merge(
    branded[[
        "fdc_id",
        "brand_owner",
        "brand_name",
        "serving_size", "serving_size_unit",
        "household_serving_fulltext",
        "gtin_upc",
    ]],
    on="fdc_id",
    how="left",
)

print("Branded-only po merge z branded:", food_b.shape)
display(food_b.head(5))
````

### makra per 100 g i merge z produktami

- wybieramy z `food_nutrient.csv` tylko makroskładniki: kalorie, białko, węgle, tłuszcz
- robimy pivot tak, żeby dla każdego `fdc_id` była jedna linia z kolumnami:
  - `kcal100g`
  - `protein100g`
  - `carbs100g`
  - `fat100g`
- łączymy te makra z tabelą `food_b` tak, aby każdy produkt branded miał komplet danych żywieniowych na 100 g
- wynik to `df_branded_macros`, czyli tabela produktów z podstawowymi informacjami i makrami


````python
# mapowanie nutrient_id -> nazwa kolumny 
NUTRIENT_MAP = {
    1008: "kcal100g",     # kcal
    1003: "protein100g",  # białko
    1005: "carbs100g",    # węglowodany
    1004: "fat100g",      # tłuszcz
}

# bierzemy tylko makra
fn_macros = food_nutrient[food_nutrient["nutrient_id"].isin(NUTRIENT_MAP.keys())]

# pivot: jedna linia per fdc_id, kolumny = kcal100g, protein100g, ...
fn_piv = (
    fn_macros.pivot_table(
        index="fdc_id",
        columns="nutrient_id",
        values="amount",
        aggfunc="first",
    )
    .rename(columns=NUTRIENT_MAP)
    .reset_index()
)

print("fn_piv (makra):", fn_piv.shape)
display(fn_piv.head(5))

# merge: produkt + makra
df_branded_macros = (
    food_b.merge(fn_piv, on="fdc_id", how="left")
          .sort_values("publication_date")
          .reset_index(drop=True)
)

print("df_branded_macros:", df_branded_macros.shape)
display(df_branded_macros.head(5))

print("\nNon-null counts (makra):")
print(df_branded_macros[["protein100g","carbs100g","fat100g","kcal100g"]].notna().sum())
````

### Jednostki porcji dla Branded Foods

W katalogu *Branded Foods* ( w przeciwieńswie do *foundational food* ) USDA **nie ma tabeli `food_portion`**, więc brak jest oficjalnych `gram_weight` dla jednostek typu *slice*, *cup*, *piece* itd. Plik `measure_unit.csv` jest tylko słownikiem nazw jednostek (ID -> nazwa), bez informacji o masie. Dlatego trzeba będzie być kreatywnym z parsowaniem `household_serving_fulltext`

W tej komórce:

- dla każdego produktu z `df_branded_macros` buduje jednostkę porcji:
  - `unit_usdaUnitId`
  - `unit_name`
  - `unit_symbol`
  - `gramsPerUnit`
- gdy `serving_size_unit` = `"g"`:
  - odczytywana jest masa porcji podana przez USDA (`serving_size`, np. 30 g)
  - parser sprawdza, czy `household_serving_fulltext` zaczyna się od wzorca typu `"X unit"` (np. `"2 pieces"`, `"1 slice"`)
  - jeśli uda się dopasować liczbę i nazwę jednostki:
      - liczba jednostek = X (np. 2)
      - `gramsPerUnit` = (`serving_size` w gramach) / X  
        przykład: 30 g / 2 = 15 g na 1 piece
      - jeśli nazwa jednostki (np. `"piece"`, `"slice"`, `"cup"`) występuje w `measure_unit.csv`, to ustawiany jest `unit_usdaUnitId`
      - jeśli jednostka nie istnieje w `measure_unit.csv`, `unit_usdaUnitId` pozostaje `None`, ale `unit_name` i `gramsPerUnit` są zachowane

- gdy `serving_size_unit` = `"ml"` lub dane są nieparsowalne / podejrzane:
  - stosuje bezpieczny fallback: `unit_name = "gram"`, `unit_symbol = "g"`, `gramsPerUnit = 1.0`, `unit_usdaUnitId = None`

Uzupełnianie brakujących `unit_usdaUnitId` ( zostanie to zrobione analogicznie jak w analizie *foundational foods* ) jest robione później, na gotowym DataFrame.


````python
import re

# przygotowanie measure_unit (tylko name/id)
mu = measure_unit.copy()
mu["name_l"] = mu["name"].str.lower().str.strip()

def find_unit_id(name: str):
    if not isinstance(name, str):
        return None
    s = name.strip().lower()
    hit = mu[mu["name_l"] == s]
    if len(hit):
        return int(hit.iloc[0]["id"])
    return None

# aliasy do parsowania jednostek z household_serving_fulltext
UNIT_ALIASES = {
    "tsp": "teaspoon",
    "teaspoons": "teaspoon",
    "tbsp": "tablespoon",
    "tablespoons": "tablespoon",
    "cup": "cup",
    "cups": "cup",
    "slice": "slice",
    "slices": "slice",
    "piece": "piece",
    "pieces": "piece",
    "serving": "serving",
    "servings": "serving",
    "container": "container",
    "containers": "container",
    "pkg": "package",
    "package": "package",
    "packages": "package",
    "g": "gram",
}

NUM_UNIT_RE = re.compile(r"^\s*(\d+(?:\.\d+)?)\s*([a-zA-Z]+)")

def parse_household(txt: str):
    
    #Zwraca (qty, unit_norm) jeśli tekst zaczyna się od 'X unit'
    #(np. '2 pieces', '1 slice'). W przeciwnym razie None.
    
    if not isinstance(txt, str):
        return None
    m = NUM_UNIT_RE.match(txt)
    if not m:
        return None
    qty = float(m.group(1))
    unit_raw = m.group(2).lower()
    unit_norm = UNIT_ALIASES.get(unit_raw, unit_raw)
    return qty, unit_norm


def build_unit_row(row):
    serving_size = row.serving_size
    su = row.serving_size_unit
    serving_unit = (su or "").strip().lower() if isinstance(su, str) else ""
    household_txt = row.household_serving_fulltext

    # domyślny fallback: gram / 1 g, ale unit_usdaUnitId = None
    unit_id = None
    unit_name = "gram"
    unit_symbol = "g"
    grams_per_unit = 1.0

    # przypadek ml -> twardy fallback
    if serving_unit in ("ml", "milliliter"):
        return pd.Series({
            "unit_usdaUnitId": unit_id,
            "unit_name": unit_name,
            "unit_symbol": unit_symbol,
            "gramsPerUnit": grams_per_unit,
        })

    # przypadek gram -> próbujemy sparsować household_serving_fulltext
    if serving_unit in ("g", "gram"):
        parsed = parse_household(household_txt)
        if parsed and pd.notna(serving_size) and serving_size > 0:
            qty, unit_txt = parsed
            if qty > 0:
                uid = find_unit_id(unit_txt)
                if uid is not None:
                    unit_id = uid   # USDA zna tę jednostkę
                else:
                    unit_id = None  # USDA nie zna, ale nazwa i tak zostaje

                unit_name = unit_txt
                unit_symbol = unit_txt
                grams_per_unit = float(serving_size) / float(qty)

    # reszta → fallback gram / 1 g

    return pd.Series({
        "unit_usdaUnitId": unit_id,
        "unit_name": unit_name,
        "unit_symbol": unit_symbol,
        "gramsPerUnit": grams_per_unit,
    })


# zastosowanie do całego DF z makrami
unit_df = df_branded_macros.apply(build_unit_row, axis=1)

# tu dokładamy też brand_owner / brand_name
df_products = pd.concat(
    [
        df_branded_macros[[
            "fdc_id",
            "description",
            "brand_owner",
            "brand_name",
            "kcal100g",
            "protein100g",
            "carbs100g",
            "fat100g",
        ]],
        unit_df,
    ],
    axis=1,
)

df_products = df_products.rename(columns={
    "fdc_id": "usdaId",
    "description": "name",
})

# tutaj powstaje nazwa "nazwa_produktu - marka"
def build_display_name(row):
    base = row["name"]
    brand = row["brand_name"]
    owner = row["brand_owner"]

    if isinstance(brand, str) and brand.strip():
        return f"{base} - {brand.strip()}"
    if isinstance(owner, str) and owner.strip():
        return f"{base} - {owner.strip()}"
    return base

df_products["displayName"] = df_products.apply(build_display_name, axis=1)

print("df_products:", df_products.shape)
display(df_products[["usdaId", "name", "brand_name", "brand_owner", "displayName"]].head(5))
````

### Dla przypomnienia - Jak powstaje `unit_name` w Branded Foods?

Branded Foods nie mają tabeli `food_portion`, więc USDA nie podaje oficjalnych jednostek ani gramatur porcji.  
Dlatego `unit_name` pochodzi z dwóch źródeł:

1. **Parsowanie `household_serving_fulltext`**  
   Jeśli tekst wygląda jak `X unit` (np. `2 slices`, `1 piece`, `3 tbsp`), parser wyciąga liczbę i nazwę jednostki, a następnie liczy:  
   `gramsPerUnit = serving_size / X`.

2. **Fallback**  
   Jeśli parsowanie się nie powiodło lub jednostka jest bez sensu, ustawiane jest:  
   `unit_name = "gram"`, `unit_symbol = "g"`, `gramsPerUnit = 1.0`, `unit_usdaUnitId = None`.


### Nadawanie sztucznych `usdaUnitId` dla niestandardowych jednostek

Branded Foods posiadają wiele jednostek, które nie występują w oficjalnej tabeli `measure_unit`
(USDA unit IDs mieszczą się w zakresie 1000–2000).  
Dla takich przypadków tworzymy własne, stabilne identyfikatory:

- `gram` zachowuje wcześniej przyjęty placeholder ( z poprzedniej obróbki danych *foundational foods`) `usdaUnitId = 2048`,
- każda inna niestandardowa jednostka dostaje nowe ID zaczynające się od **3000**,
- numery są nadawane deterministycznie, według sortowanej listy unikalnych `unit_name`.

Dzięki temu unikamy kolizji z USDA i możemy później łatwo zarządzać dodatkowymi jednostkami.

Na początek zajmiemy się tą specjalną, niepoliczalną jednostką "gram". Wypiszemy wszystkie rekordy które jako `unit_name` mają jakąś jednostke wagi, także "amerykańską" ( np uncja ) oraz wszystkie te, co się zaczynają na `g`

````python
import re

all_units = sorted(set(df_products["unit_name"].dropna()))

# wzorce jednostek wagi
weight_regex = re.compile(
    r"^(g|gm|gr|gram|grm|gms|ounce|oz|lb|lbs|pound|pounds|kg|kilogram)", 
    flags=re.IGNORECASE
)

weight_like_units = [u for u in all_units if weight_regex.match(u)]

print("Jednostki wyglądające na wagę:")
for u in weight_like_units:
    print(" -", u)

print("\nŁącznie:", len(weight_like_units))
````

#### Normalizacja pseudojednostek wagi

Branded Foods zawiera wiele *"dziwnych"* wariantów typu *grm*, *gm*, *ozs*, *oza*, *ounce*, które w praktyce oznaczają to samo.  
Nie mają one wiarygodnych gramWeight, więc wszystkie normalizujemy do jednej jednostki: **gram (ID 2048, 1 g)**.  

````python
weight_like_units = {
    "gm", "gr", "grm", "gram", "grams",
    "oz", "oza", "ozs", "ounce", "ounces"
}

def normalize_weight_units(row):
    name = str(row["unit_name"]).lower().strip()

    if name in weight_like_units:
        return pd.Series({
            "unit_name": "gram",
            "unit_symbol": "g",
            "unit_usdaUnitId": 2048,
            "gramsPerUnit": 1.0
        })
    
    # wpp zwracamy oryginał
    return pd.Series({
        "unit_name": row["unit_name"],
        "unit_symbol": row["unit_symbol"],
        "unit_usdaUnitId": row["unit_usdaUnitId"],
        "gramsPerUnit": row["gramsPerUnit"]
    })

df_products[["unit_name", "unit_symbol", "unit_usdaUnitId", "gramsPerUnit"]] = \
    df_products.apply(normalize_weight_units, axis=1)

print("Normalizacja jednostek wagi wykonana.")
````

Ogólnie można zauważyć, że pojawiają się `unit_name` co mają 1 lub 2 znaki długosci. Zakładamy, że unit typu `x` albo `th` nie wnosi nam żadnej wartościowej informacji, więc tak samo, sprrowadzimy je do fall-back unit: czyli grama

````python
# Zamieniamy na string, żeby bezpiecznie liczyć długości
unit_names = df_products["unit_name"].astype("string")

mask_null = unit_names.isna()

mask_len1 = unit_names.str.len() == 1

mask_len2 = unit_names.str.len() == 2

print("Null unit_name:", mask_null.sum())
print("Length 1:", mask_len1.sum())
print("Length 2:", mask_len2.sum())

print("\nPrzykłady (len = 1):")
print(unit_names[mask_len1].dropna().unique()[:20])

print("\nPrzykłady (len = 2):")
print(unit_names[mask_len2].dropna().unique()[:20])
````

````python
# Normalizacja jednostek o długości 0/1/2 znaków lub null
def normalize_short_units(row):
    name = row["unit_name"]

    if pd.isna(name) or len(str(name)) <= 2:
        return pd.Series({
            "unit_name": "gram",
            "unit_symbol": "g",
            "unit_usdaUnitId": 2048,
            "gramsPerUnit": 1.0
        })

    return pd.Series({
        "unit_name": row["unit_name"],
        "unit_symbol": row["unit_symbol"],
        "unit_usdaUnitId": row["unit_usdaUnitId"],
        "gramsPerUnit": row["gramsPerUnit"]
    })

df_products[["unit_name", "unit_symbol", "unit_usdaUnitId", "gramsPerUnit"]] = \
    df_products.apply(normalize_short_units, axis=1)

print("Normalizacja jednostek ≤2 znaki wykonana.")
````

Teraz dla pozostałych jednostek zrobimy "customowe" UsdaUnitID, zaczynające się od 3000, aby w razie czego możnaby było je łatwo odróżnić. Unit name i unit symbol będą identyczne. 

````python
# Customowe usdaUnitId dla pozostałych jednostek (tam, gdzie unit_usdaUnitId jest null)

mask_custom = df_products["unit_usdaUnitId"].isna()

custom_unit_names = (
    df_products.loc[mask_custom, "unit_name"]
    .dropna()
    .sort_values()
    .unique()
)

print("Rekordów bez usdaUnitId:", mask_custom.sum())
print("Unikalnych unit_name w tej grupie:", len(custom_unit_names))

BASE_ID = 3000
name_to_id = {name: BASE_ID + idx for idx, name in enumerate(custom_unit_names)}

# przypisujemy nowe ID tylko tam, gdzie wcześniej było NaN
df_products.loc[mask_custom, "unit_usdaUnitId"] = (
    df_products.loc[mask_custom, "unit_name"].map(name_to_id)
)

# dla customowych jednostek symbol = nazwa
df_products.loc[mask_custom, "unit_symbol"] = df_products.loc[mask_custom, "unit_name"]

print("Przykładowe customowe jednostki:")
display(
    df_products.loc[mask_custom, ["unit_name", "unit_symbol", "unit_usdaUnitId"]]
              .drop_duplicates()
              .head(20)
)
````

### Wykrywanie popularnych marek na podstawie brand_name / brand_owner

- Tworzona jest robocza kopia `df_products` (`tmp`) oraz pomocnicze kolumny `brand_name_l` i `brand_owner_l` z nazwami marek i właścicieli w formie znormalizowanej (lowercase, string).
- Słownik `POPULAR_BRAND_PATTERNS` definiuje wzorce tekstowe dla wybranych popularnych marek (np. `monster`, `coca_cola`, `kinder`, `lays`).
- Funkcja `detect_popular_brand` łączy znormalizowane `brand_name_l` i `brand_owner_l`, a następnie sprawdza, czy któryś z wzorców marek występuje w tym tekście; zwraca odpowiedni tag (np. `"monster"`) lub `None`.
- Dla każdego produktu wyznaczany jest tag `popularBrandTag`, a następnie zliczana jest liczba produktów przypadających na poszczególne marki oraz łączna liczba produktów popularnych marek.
- Na końcu wyświetlana jest próbka (do 50 wierszy) produktów, dla których `popularBrandTag` jest niepuste, wraz z podstawowymi informacjami o nazwie i marce.


````python
import pandas as pd

# kopia robocza do pracy z markami
tmp = df_products.copy()

# normalizacja brand_name / brand_owner
tmp["brand_name_l"] = tmp["brand_name"].astype("string").str.lower().fillna("")
tmp["brand_owner_l"] = tmp["brand_owner"].astype("string").str.lower().fillna("")

POPULAR_BRAND_PATTERNS = {
    "nestle": ["nestle", "nestlé"],
    "kelloggs": ["kellogg", "kellogg's"],
    "heinz": ["heinz"],
    "danone": ["danone"],
    "pepsi": ["pepsi"],
    "mars": [" mars "], 
    "coca_cola": ["coca-cola"],
    "lipton": ["lipton"],
    "oreo": ["oreo"],
    "lays": ["lay's", "lays"],
    "monster": ["monster energy", "monster beverage"],
    "pringles": ["pringles"],
    "doritos": ["doritos"],
    "cheetos": ["cheetos"],
    "kinder": ["kinder"],
    "haribo": ["haribo"],
    "red_bull": ["red bull"],
    "kitkat": ["kit kat", "kitkat"],
    "knorr": ["knorr"],
    "fanta": ["fanta"],
    "milka": ["milka"],
    "hellmanns": ["hellmann", "hellmann's"],
    "dr_oetker": ["dr. oetker", "dr oetker"],
    "snickers": ["snickers"],
    "twix": ["twix"],
    "nutella": ["nutella"],
    "nescafe": ["nescafe", "nescafé"],
    "activia": ["activia"],
    "sprite": ["sprite"],
    "nesquik": ["nesquik"],
}

def detect_popular_brand(row):
    text = (row["brand_name_l"] + " " + row["brand_owner_l"]).strip()
    if not text:
        return None
    for tag, patterns in POPULAR_BRAND_PATTERNS.items():
        for pat in patterns:
            if pat in text:
                return tag
    return None

tmp["popularBrandTag"] = tmp.apply(detect_popular_brand, axis=1)

brand_counts = tmp["popularBrandTag"].value_counts().sort_values(ascending=False)

print("Liczba produktów dla poszczególnych marek:")
print(brand_counts)

print("\nŁącznie produktów popularnych marek:",
      tmp["popularBrandTag"].notna().sum())

display(
    tmp.loc[tmp["popularBrandTag"].notna(),
            ["usdaId", "name", "brand_name", "brand_owner", "popularBrandTag"]]
       .head(30)
)
````

### Selekcja produktów najpopularniejszych marek

- Definiowana jest lista `TOP_BRANDS` zawierająca wybrane popularne marki (zgodne ze znormalizowanymi tagami z `popularBrandTag`).
- Z `tmp` wybierane są tylko te produkty, które mają `popularBrandTag` należący do `TOP_BRANDS`.
- Dla każdej marki pobieranych jest maksymalnie 20 produktów (pierwsze wystąpienia), aby uzyskać reprezentatywne, ale ograniczone próbki.
- Wybrane fragmenty są łączone w jeden DataFrame (`small_df`), następnie deduplikowane po `usdaId` i resetowany jest indeks.


````python
import pandas as pd

TOP_BRANDS = [
    "nestle",
    "kelloggs",
    "heinz",
    "danone",
    "pepsi",
    "mars",
    "coca_cola",
    "lipton",
    "oreo",
    "lays",
    "monster",
    "pringles",
    "doritos",
    "cheetos",
    "kinder",
    "haribo",
    "red_bull",
    "kitkat",
    "knorr",
    "fanta",
    "milka",
    "hellmanns",
    "dr_oetker",
    "snickers",
    "twix",
    "nutella",
    "nescafe",
    "activia",
    "sprite",
    "nesquik",
]

popular = tmp[tmp["popularBrandTag"].isin(TOP_BRANDS)].copy()

chunks = []
for tag in TOP_BRANDS:
    sub = popular[popular["popularBrandTag"] == tag].head(20)
    chunks.append(sub)

small_df = (
    pd.concat(chunks, ignore_index=True)
      .drop_duplicates(subset=["usdaId"])
      .reset_index(drop=True)
)

print("small_df shape (po wyborze marek):", small_df.shape)
display(small_df[["usdaId", "name", "brand_name", "brand_owner", "popularBrandTag"]].head(30))
````

### Finalna obróbka do json'a

Dodatkowo przeglądając mniejszą liczbe rekordów, można było zauważyć duplikaty, np *Keebler Cones Waffle Cone 216ct*. 
Duplikaty zostaną usuniete, dodatkowo finalna nazwa produktu to będZie `name` + `popular brand tag`, ponieważ będzie to krótsze oraz dość jednoznacze, do jakiej marki należy produkt. Finalnie jeszcze nazwa będzie w *lowercase*, ponieważ wszytskie nazwy produktów w *branded foods* USDA postanowiło dać samymi dużymi literami, a to wygląda nieładnie oraz nie jest spójne z obrobionymi produktami z *foundational foods*.  

Poniżek: 
- Dla każdego rekordu tworzona jest nazwa `jsonName` w formacie  
  `"nazwa_produktu - brandtag"` (całość w lowercase).  
  Jeśli produkt nie ma `popularBrandTag`, używana jest sama nazwa.
- Dane są sortowane po `usdaId`, a następnie usuwane są duplikaty po `jsonName`, tak aby każdy produkt miał unikalną nazwę.
- Sprawdzana jest obecność wymaganych kolumn (m.in. makra i informacje o jednostce). Brak którejkolwiek powoduje błąd.
- Każdy wiersz jest konwertowany do finalnej struktury JSON:  
  `usdaId`, `name`, sekcja `unit`, masa jednostki i makroskładniki na 100 g.
- Wszystkie rekordy są zapisywane do pliku `branded_foods_small.json`.


````python
import json
from pathlib import Path

# budujemy nazwę: "nazwa_produktu - brandtag", wszystko lowercase
def build_json_name(row):
    base = str(row["name"]).lower().strip()
    tag = str(row.get("popularBrandTag", "") or "").lower().strip()
    tag_pretty = tag.replace("_", " ") if tag else ""
    if tag_pretty:
        return f"{base} - {tag_pretty}"
    return base

small_df["jsonName"] = small_df.apply(build_json_name, axis=1)

# usuwamy duplikaty po jsonName, zostawiamy rekord z najmniejszym usdaId
small_df = (
    small_df.sort_values("usdaId")
            .drop_duplicates(subset=["jsonName"])
            .reset_index(drop=True)
)

print("Po deduplikacji po jsonName:", small_df.shape)

# walidacja wymaganych kolumn
needed_cols = [
    "usdaId",
    "jsonName",
    "unit_usdaUnitId",
    "unit_name",
    "unit_symbol",
    "gramsPerUnit",
    "kcal100g",
    "protein100g",
    "carbs100g",
    "fat100g",
]

missing = set(needed_cols) - set(small_df.columns)
if missing:
    raise ValueError(f"Brakuje kolumn w small_df: {missing}")

def row_to_record_small(row):
    usda_unit_id = row["unit_usdaUnitId"]
    if pd.isna(usda_unit_id):
        usda_unit_id = None
    else:
        usda_unit_id = int(usda_unit_id)

    return {
        "usdaId": int(row["usdaId"]),
        "name": row["jsonName"],  
        "unit": {
            "usdaUnitId": usda_unit_id,
            "name": str(row["unit_name"]),
            "symbol": str(row["unit_symbol"]),
        },
        "gramsPerUnit": float(row["gramsPerUnit"]),
        "kcal100g": float(row["kcal100g"]),
        "protein100g": float(row["protein100g"]),
        "carbs100g": float(row["carbs100g"]),
        "fat100g": float(row["fat100g"]),
    }

records_small = [row_to_record_small(row) for _, row in small_df.iterrows()]
print("Liczba rekordów w branded_foods_small:", len(records_small))

OUTPUT_SMALL = Path(r"C:\Users\piotr\dieti_data\dYeti_data\jowisz_2\branded_foods_small.json")
with OUTPUT_SMALL.open("w", encoding="utf-8") as f:
    json.dump(records_small, f, ensure_ascii=False, indent=2)

print("Zapisano:", OUTPUT_SMALL)
````
